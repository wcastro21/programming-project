## Imported libraries and .csv variable

```python
import matplotlib.pyplot as plt                 #matplotlib library allows us to graph our findings
import pandas as pd                             # pandas library allows us to open and navigate .csv

df = pd.read_csv('data/fraud_detection.csv')    #creates a variable that allows us to quickly navigate .csv file
```

### Question 1: How many fraudalent purchases were there compared to real purchases?

```python
fraud_count = (df['label'] == 1).value_counts()  # variable filters the data to only show fraudalent purchases 

mylabels = ['Real', 'Fraudalent']   # creates labels for pie chart
plt.figure(figsize=(6,6))           # makes pie chart bigger
plt.pie(fraud_count, labels = mylabels, autopct='%1.1f%%')    #creates pie chart
plt.title('Fraudalent purchases')   # creates title for pie chart
plt.legend()    # creates legend for pie chart
plt.show()  # shows the graph

print(fraud_count)    # prints the actual amount of fraudalent purchases
```

Answer: Out of 1000 purchases, only 50 of them (or 5%) were fraudalent. 

### Question 2: What purchase type was most prevalent in these Fraud Transactions?

```python
fraud_transactions = df[df['label'] == 1]       # global variable that filters the data to only show fraudalent purchases

purchase_type = fraud_transactions['merchant_type'].value_counts()  # variable filters the data to only show fraudalent purchases by purchase type
purchase_type.plot(kind='bar') # creates bar chart

## labels for bar chart
plt.title('Fraudalent purchases by Purchase Type')   # creates title for bar chart
plt.xlabel('Purchase Type')   # creates x axis label for bar chart
plt.ylabel('Number of Fraudalent Purchases')   # creates y axis label for bar chart
plt.xticks(rotation=45)   # rotates x axis labels for better readability
plt.show()

print(purchase_type)    # prints the actual amount of fraudalent purchases by purchase type
```

Answer: While the data set doesn't give us a specific type of purchase type, the "others" category is the most prevalent purchase type.

### Question 5: Which Device Type on average spent the most amount of money during these Fraud Transactions?

```python
average_device_type = fraud_transactions.groupby('device_type')['amount'].mean()      # variable calculates the average amount each device spent
average_device_type.plot(kind='bar')    # creates a bar graph of the average amount spent by each device

plt.title('Average Amount Spent by Device Type')    # creates title for bar graph
plt.xlabel('Device Type')    # creates x-axis label for bar graph
plt.ylabel('Average Amount Spent')    # creates y-axis label for bar graph
plt.xticks(rotation=45)    # rotates x-axis labels for better readability
plt.show()    # shows the bar graph

print(average_device_type)   # prints the average amount spent by each device type
```

Answer: The Device Type that has the highest average amount spent for Fraudalent purchases are Tablets, where the average purchase is $114.57.
