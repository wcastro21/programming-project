## Imported libraries and .csv variable

```python
import matplotlib.pyplot as plt                 #matplotlib library allows us to graph our findings
import pandas as pd                             # pandas library allows us to open and navigate .csv

df = pd.read_csv('data/fraud_detection.csv')    #creates a variable that allows us to quickly navigate .csv file
```

### Question 1: How much of each purchase type was used in these Fraud Transactions?

```python
merchant_type = df['merchant_type'].value_counts()  # variable counts amount of purchases for each purchase type
merchant_type.plot(kind='bar')      # creates a bar chart

## labels for graph
plt.title('Amount of purchases for every Fraud purchase type')      # creates title for graph
plt.xlabel('Fraud Purchase Type')       # labels x axis for graph
plt.ylabel('Amount of purchases')       # labels y axis for graph
plt.xticks(rotation=45)                 # rotates fraud purchase type labels to make it easier to read
plt.show()                              # shows the graph

print(merchant_type)    #prints the value counts
```

### Question 4: Which Device Type on average spent the most amount of money during these Fraud Transactions?

```python
average_device_type = df.groupby('device_type')['amount'].mean()        # variable calculates the average amount each device spent
average_device_type.plot(kind='bar')        # plots the variable as a bar graph

## labels for graph
plt.title('Average amount spent on Fraud Transactions for each Device Type')    # creates title label for graph
plt.xlabel('Devices')       # creates x label for graph
plt.ylabel('Average Amount')        # creates y label for graph
plt.xticks(rotation=45)     # rotates x ticks to make graph easier to read
plt.show()          # shows graph

print(average_device_type)  # prints the averages
```

Although the averages are very close, scammers are spending more on fraud transactions using Tablets.
